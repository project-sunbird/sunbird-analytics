```python
import json
data = []
with open('DeltaDump2015All.txt') as f:
    for line in f:
        data.append(json.loads(line))
d=dict()
for item in data:
    if(item['uid']!="none" and item['uid'] not in d.keys()):
        ud=item['uid']
        c=["No"]
        t=list()
        for time in data:
            if(time['uid']==ud and'id' in time['edata']['eks'].keys()):
                s=time['edata']['eks']['id']
                if("suitcase" in s):
                    c[0]="Yes"
            if(time['uid']==ud):
                t.append(time['ts'])
        sorted(t)
        hr=t[0]
        c.append(int(hr[11:13]))
        d[ud]=c
print "{:<38}{:<24}{:<12}".format('UID', "Access rewards Page", "Hour of Day")
for k, v in d.iteritems():
    reward, ti = v
    print "{:<38}{:<24}{:<12}".format(k, reward, ti)
```
